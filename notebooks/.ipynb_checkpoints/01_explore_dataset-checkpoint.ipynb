# 01 — Explore Biomedical NER Datasets (JNLPBA & BC5CDR)

This notebook is for inspect the datasets for fine‑tune BioBERT.

- Load **JNLPBA** (genes/proteins) and **BC5CDR** (chemicals/diseases)
- Map label IDs → names
- Check label distribution and sequence lengths
- Peek at a few annotated examples


```python
# If running in Colab, uncomment the next line to ensure required libraries are present
# !pip -q install datasets>=2.19

from datasets import load_dataset
from collections import Counter
import itertools
from statistics import median

def load_ds(name: str):
    ds = load_dataset(name)
    labels = ds["train"].features["ner_tags"].feature.names
    id2label = {i: n for i, n in enumerate(labels)}
    label2id = {n: i for i, n in enumerate(labels)}
    return ds, labels, id2label, label2id

print("Ready.")
```

## 1) Load JNLPBA and inspect a sample

```python
DATASET = "jnlpba"  # change to 'bioasq/BC5CDR' for BC5CDR
ds, labels, id2label, label2id = load_ds(DATASET)

print(ds)  #  train/validation/test sizes
ex = ds["train"][0]
print("Columns:", list(ex.keys()))
print("Tokens (first 40):", ex["tokens"][:40])
print("NER tags (IDs, first 40):", ex["ner_tags"][:40])
print("Labels:", labels)

decoded = [id2label[i] for i in ex["ner_tags"][:40]]
print("Decoded tags (first 40):", decoded)
```

## 2) Label distribution per split (class balance)

```python
def label_distribution(split="train"):
    ctr = Counter(itertools.chain.from_iterable(ds[split]["ner_tags"]))
    total = sum(ctr.values())
    rows = [(id2label[i], c, round(100*c/total, 2)) for i, c in ctr.most_common()]
    return rows

for split in ["train", "validation", "test"]:
    print(f"\n== {split.upper()} ==")
    for name, count, pct in label_distribution(split):
        print(f"{name:12s} {count:7d}  {pct:5.2f}%")
```

## 3) Sequence length stats (choose a safe `max_length`)

```python
lengths = [len(x) for x in ds["train"]["tokens"]]
lengths_sorted = sorted(lengths)
p95 = lengths_sorted[int(0.95*len(lengths_sorted))]
print({
    "min": min(lengths),
    "median": median(lengths),
    "p95": p95,
    "max": max(lengths),
})
print("Recommended max_length:", 256 if p95 <= 256 else 320 if p95 <= 320 else 384)
```

## 4) Visual check: show tokens with BIO labels

```python
def show_example(idx=0, split="train", max_tokens=200):
    toks = ds[split][idx]["tokens"]
    tags = [id2label[i] for i in ds[split][idx]["ner_tags"]]
    pairs = [f"{t}/{g}" for t, g in zip(toks, tags)]
    print(" ".join(pairs[:max_tokens]))

show_example(0, "train")
```

## 5) Entity mention counts per split (count B- tags)

```python
from collections import Counter

def entity_counts(split="train"):
    ctr = Counter()
    for tag_seq in ds[split]["ner_tags"]:
        for t in tag_seq:
            name = id2label[t]
            if name.startswith("B-"):
                ctr[name[2:]] += 1
    return ctr

for split in ["train", "validation", "test"]:
    print(split, entity_counts(split))
```
