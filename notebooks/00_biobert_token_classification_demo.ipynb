# Token Classification with BioBERT — A Worked Example

## 1. Input Sentence

We start with a biomedical sentence:

```
"Tranexamic acid prevents intravascular coagulation."
```

Our goal: **detect entities** (chemicals, diseases) at the **token level**.

## 2. Tokenization

Using BioBERT’s WordPiece tokenizer:

```
["[CLS]", "Tran", "##ex", "##amic", "acid", "prevents", "intravascular", "coagulation", ".", "[SEP]"]
```

Notice:
- `"Tranexamic"` was split into subwords: `["Tran", "##ex", "##amic"]`.
- Each subword will receive a label.

## 3. BIO Labels

We use the **BIO scheme** (Begin, Inside, Outside) for entity tagging.  
- **Chemical**: "Tranexamic acid"  
- **Disease**: "intravascular coagulation"  

| Token        | Label        |
|--------------|--------------|
| [CLS]        | O            |
| Tran         | B-Chemical   |
| ##ex         | I-Chemical   |
| ##amic       | I-Chemical   |
| acid         | I-Chemical   |
| prevents     | O            |
| intravascular| B-Disease    |
| coagulation  | I-Disease    |
| .            | O            |
| [SEP]        | O            |


## 4. Model Forward Pass (Conceptual)

We pass the tokens through **BioBERT + token classification head**:

```python
inputs = tokenizer(sentence)
logits = model(inputs)   # shape: [batch, seq_len, num_labels]
preds = argmax(logits)   # predicted label IDs
```

- `logits`: raw scores per token per label.  
- `preds`: best label for each token.

## 5. Map Predictions Back to Words

We map prediction IDs → label names and reconstruct spans.

Example (ideal prediction):

```
Tran            → B-Chemical
##ex            → I-Chemical
##amic          → I-Chemical
acid            → I-Chemical
prevents        → O
intravascular   → B-Disease
coagulation     → I-Disease
.               → O
```

## ✅ Conclusion

- **Token classification** = predict a label per token.  
- **BIO tagging** allows span reconstruction into entities.  
- In this example:
  - `"Tranexamic acid"` → Chemical
  - `"intravascular coagulation"` → Disease  

This is the foundation for **biomarker extraction** using BioBERT.
